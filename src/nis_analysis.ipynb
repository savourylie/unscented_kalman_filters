```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def ukf_csv_to_arr(filename):
    df = pd.read_csv(filename, header=None)
    arr = df.values.reshape(df.size)
    arr = np.array([x for x in arr.tolist() if x is not None])
    
    return arr
```

```python
lidar = ukf_csv_to_arr('nis_lidar.csv')
radar = ukf_csv_to_arr('nis_radar.csv')
```

```python
pd.Series(lidar).plot()
plt.axhline(y=5.991, linewidth=2, color = 'k')
plt.show()

pd.Series(radar).plot()
plt.axhline(y=7.815, linewidth=2, color = 'k')
plt.show()
```
